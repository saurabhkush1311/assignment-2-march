```python
#Question2:
#Answer:
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Scatter-Plot")
plt.show()
```

```python
#Question3:
#Answer:

#The subplot() function is used to create multiple plots within a single figure

x =  np.array([0,1,2,3,4,5])
y = np.array([0, 100, 200, 300, 400, 500])
y1 = np.array([50, 20, 40, 20, 60, 70])
y2 = np.array([10, 20, 30, 40, 50, 60])
y3 = np.array([200, 350, 250, 550, 450, 150])

plt.subplot(2,2,1)
plt.plot(x,y,label = "line 1")
plt.legend()

plt.subplot(2,2,2)
plt.plot(x,y2,label = "line 2")
plt.legend()

plt.subplot(2,2,3)
plt.plot(x,y2,label = "line 3")
plt.legend()

plt.subplot(2,2,4)
plt.plot(x,y3,label = "line 4")
plt.legend()

plt.tight_layout()
plt.show()
```

```python
#Question4:
#Answer:

#A bar plot (or bar chart) is a type of data visualization that represents categorical data using rectangular bars.
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company,profit)
plt.show()
```

```python
plt.barh(company,profit)
plt.show()
```

```python
#Question5:
#Answer:
"""A box plot, also known as a box-and-whisker plot, is a type of data visualization that displays the distribution of a dataset
along with key statistical measures such as the median, quartiles, and potential outliers.
The box plot is particularly useful for gaining insights into the spread, skewness, and presence of outliers in a dataset."""
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
data = [box1,box2]
plt.boxplot(data)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("box-plot")
plt.show()
```
